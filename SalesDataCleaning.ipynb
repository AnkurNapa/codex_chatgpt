# Sales Data Cleaning with PySpark
This notebook demonstrates data cleaning steps for sales data.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("SalesCleaning").getOrCreate()

# Load sales data from CSV file
data = spark.read.csv('sales_data.csv', header=True, inferSchema=True)

# Drop duplicates and null rows
data_clean = data.dropDuplicates().na.drop()

# Separate open and closed orders
open_orders = data_clean.filter(col('status') == 'open')
closed_orders = data_clean.filter(col('status') == 'closed')

open_orders.show()
closed_orders.show()
```
